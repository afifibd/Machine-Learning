```python
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih kolom fitur numerik
numerical_features = ['Jumlah Tindak Pidana 2021', 'Jumlah Tindak Pidana 2022',
                      'Penyelesaian tindak pidana 2021(%)', 'Penyelesaian tindak pidana 2022(%)']

# Membuat boxplot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=feature, data=df, palette='pastel')
    plt.title(f'Boxplot of {feature}', fontsize=15)
    plt.xlabel(feature, fontsize=12)
    plt.show()
```

#Scatter Plot

ChatGPT Prompt : Beri kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Memilih kolom fitur numerik
numerical_features = ['Jumlah Tindak Pidana 2021', 'Jumlah Tindak Pidana 2022',
                      'Penyelesaian tindak pidana 2021(%)', 'Penyelesaian tindak pidana 2022(%)']

# Membuat pairplot untuk setiap kombinasi fitur numerik
sns.set(style="ticks")
sns.pairplot(df[numerical_features])
plt.show()
```

#Heat Map

ChatGPT Prompt : Beri kode untuk menampilkan heat map dari fitur-fitur numerik

```python
# Memilih kolom fitur numerik
numerical_features = ['Jumlah Tindak Pidana 2021', 'Jumlah Tindak Pidana 2022',
                      'Penyelesaian tindak pidana 2021(%)', 'Penyelesaian tindak pidana 2022(%)']

# Menghitung korelasi antar fitur numerik
correlation_matrix = df[numerical_features].corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Numeric Features Correlation', fontsize=15)
plt.show()
```

#Violin Plot

Prompt ChatGPT : beri kode untuk menampilkan violin plot dari fitur-fitur numerik

```python
# Memilih kolom fitur numerik
numerical_features = ['Jumlah Tindak Pidana 2021', 'Jumlah Tindak Pidana 2022',
                      'Penyelesaian tindak pidana 2021(%)', 'Penyelesaian tindak pidana 2022(%)']

# Membuat violin plot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.violinplot(y=feature, data=df, palette='pastel')
    plt.title(f'Violin Plot of {feature}', fontsize=15)
    plt.ylabel(feature, fontsize=12)
    plt.show()
```

#Distributions

Prompt ChatGPT : beri kode untuk menampilkan visualisasi Distributions

```python
# Memilih kolom fitur numerik
numerical_features = ['Jumlah Tindak Pidana 2021', 'Jumlah Tindak Pidana 2022',
                      'Penyelesaian tindak pidana 2021(%)', 'Penyelesaian tindak pidana 2022(%)']

# Membuat visualisasi distribusi untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], kde=True, color='skyblue', bins=20)
    plt.title(f'Distribution of {feature}', fontsize=15)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.show()
```

#Line Plot

Prompt ChatGPT : beri kode untuk menampilkan Line plot dari fitur-fitur numerik

```python
# Memilih kolom fitur numerik
numerical_features = ['Jumlah Tindak Pidana 2021', 'Jumlah Tindak Pidana 2022',
                      'Penyelesaian tindak pidana 2021(%)', 'Penyelesaian tindak pidana 2022(%)']

# Membuat line plot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df[feature], marker='o', linestyle='-', color='b')
    plt.title(f'Line Plot of {feature}', fontsize=15)
    plt.xlabel('Index', fontsize=12)
    plt.ylabel(feature, fontsize=12)
    plt.grid(True)
    plt.show()
```

#Bar Plot

Prompt ChatGPT : beri kode untuk menampilkan Bar plot dari fitur-fitur numerik

```python
# Memilih kolom fitur numerik
numerical_features = ['Jumlah Tindak Pidana 2021', 'Jumlah Tindak Pidana 2022',
                      'Penyelesaian tindak pidana 2021(%)', 'Penyelesaian tindak pidana 2022(%)']

# Membuat bar plot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.barplot(x=df.index, y=feature, data=df, palette='pastel')
    plt.title(f'Bar Plot of {feature}', fontsize=15)
    plt.xlabel('Index', fontsize=12)
    plt.ylabel(feature, fontsize=12)
    plt.xticks(rotation=45)
    plt.show()
```
