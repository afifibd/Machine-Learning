```python
import matplotlib.pyplot as plt
import seaborn as sns

# Memilih kolom fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat boxplot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=feature, data=df)
    plt.title(f'Boxplot of {feature}', fontsize=15)
    plt.xlabel(feature, fontsize=12)
    plt.show()
```

#Scatter Plot

ChatGPT Prompt : Beri kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Memilih kolom fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat pairplot untuk setiap kombinasi fitur numerik
sns.set(style="ticks")
sns.pairplot(df[numerical_features])
plt.show()
```

#Heat Map

ChatGPT Prompt : Beri kode untuk menampilkan heat map dari fitur-fitur numerik

```python
# Menghitung korelasi antara fitur-fitur numerik
correlation_matrix = df.corr()

# Membuat heat map dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Numeric Features Correlation', fontsize=15)
plt.show()
```

#Bar Plot

ChatGPT Prompt : Beri kode untuk menampilkan Bar Plot dari fitur-fitur numerik

```python
# Memilih hanya kolom-kolom numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=numerical_features, orient='h')
plt.title("Bar Plot untuk Fitur-Fitur Numerik")
plt.xlabel("Nilai")
plt.ylabel("Fitur")
plt.show()
```

#Line Plot

ChatGPT Prompt : Beri kode untuk menampilkan Line Plot dari fitur-fitur numerik

```python
# Memilih hanya kolom-kolom numerik
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Membuat line plot
plt.figure(figsize=(12, 8))
for column in numeric_cols.columns:
    plt.plot(df.index, df[column], label=column)
plt.title("Line Plot untuk Fitur-Fitur Numerik")
plt.xlabel("Indeks Data")
plt.ylabel("Nilai")
plt.legend()
plt.show()
```
