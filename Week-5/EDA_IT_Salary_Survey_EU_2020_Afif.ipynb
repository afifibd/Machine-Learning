ChatGPT Prompt : Beri petunjuk untuk melakukan Exploratory Data Analysis

Exploratory Data Analysis (EDA) adalah proses untuk memahami dataset Anda dengan melakukan analisis awal dan visualisasi data. Berikut adalah langkah-langkah umum yang dapat Anda ikuti untuk melakukan EDA:

1. **Memahami Data**:
   - Baca dokumentasi dataset (jika ada) untuk memahami struktur dan konten data.
   - Tampilkan beberapa baris pertama dataset menggunakan `head()` untuk melihat format dan struktur data.
   - Gunakan `info()` untuk melihat tipe data dan jumlah nilai non-null dalam setiap kolom.

2. **Pemrosesan Data Awal**:
   - Identifikasi dan tangani nilai yang hilang atau yang tidak valid.
   - Ubah tipe data jika diperlukan (misalnya, tanggal ke tipe data datetime).
   - Lakukan pemrosesan teks atau pembersihan data lainnya jika diperlukan.

3. **Statistik Deskriptif**:
   - Hitung statistik deskriptif seperti mean, median, dan standar deviasi untuk setiap kolom numerik menggunakan `describe()`.
   - Identifikasi nilai ekstrim, nilai yang hilang, atau anomali lainnya.

4. **Visualisasi Data**:
   - Gunakan plot seperti histogram, box plot, dan scatter plot untuk memahami distribusi dan hubungan antar variabel.
   - Gunakan heatmap untuk memahami korelasi antar variabel numerik.
   - Buat visualisasi tambahan yang sesuai dengan jenis data Anda, seperti diagram batang untuk data kategorikal atau time series plot untuk data deret waktu.

5. **Eksplorasi Lebih Lanjut**:
   - Identifikasi pola atau tren menarik dalam data.
   - Lakukan analisis univariat dan multivariat untuk memahami hubungan antar variabel.
   - Jelajahi pertanyaan atau hipotesis awal Anda dengan menggunakan visualisasi dan analisis statistik.

6. **Ringkasan dan Kesimpulan**:
   - Ringkas temuan utama Anda.
   - Berikan interpretasi atas temuan Anda.
   - Identifikasi pertanyaan baru atau arah penelitian yang mungkin.

Selama proses EDA, Anda juga dapat melakukan iterasi, kembali ke langkah-langkah sebelumnya, atau mengeksplorasi data lebih lanjut seiring dengan memahami lebih dalam tentang dataset Anda. Pemahaman yang baik tentang data adalah kunci untuk menghasilkan wawasan yang berharga dari analisis data.

ChatGPT Prompt : Beri kode untuk menampilkan pie diagram chart untuk fitur Age dengan min 20, max 69

```python
import matplotlib.pyplot as plt

# Diskritisasi usia menjadi beberapa kelompok
bins = [20, 30, 40, 50, 60, 70]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69']
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Hitung jumlah pengamatan dalam setiap kelompok usia
age_counts = df['Age_group'].value_counts()

# Buat pie chart
plt.figure(figsize=(8, 8))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Kelompok Usia')
plt.axis('equal')  # Memastikan pie chart berbentuk lingkaran
plt.show()
```

ChatGPT Prompt : beri kode untuk menampilkan baris data yang memiliki Age "51-60" dan "60+"

```python
# Memilih baris data yang memiliki nilai "51-60" atau "60+" pada kolom "Age_group"
age_51_60 = df.loc[df['Age_group'] == '50-59']
age_60_plus = df.loc[df['Age_group'] == '60-69']

# Menggabungkan hasil pemilihan baris
baris_data = pd.concat([age_51_60, age_60_plus])

# Menampilkan baris data yang telah dipilih
baris_data
```

ChatGPT Prompt : buatkan kode untuk visualisasi histogram, box plot, scatter plot untuk memahami distribusi dan hubungan antar variabe

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Total experience'], bins=20, kde=True)
plt.title('Distribusi Total Pengalaman')
plt.xlabel('Total Years of Experience')
plt.ylabel('Frequency')
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Experience in Germany', y='Yearly bonus and stocks', data=df)
plt.title('Hubungan antara Pengalaman di Jerman dan Bonus Tahunan + Saham')
plt.xlabel('Experience in Germany')
plt.ylabel('Yearly Bonus + Stocks (EUR)')
plt.show()

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total experience', y='Yearly brutto salary', data=df)
plt.title('Hubungan antara Total Pengalaman dan Gaji Bruto Tahunan')
plt.xlabel('Total Years of Experience')
plt.ylabel('Yearly Brutto Salary (EUR)')
plt.show()
```

 ChatGPT Prompt : buatkan heatmap untuk memvisualisasikan korelasinya antar kolom

```python
# Memilih hanya kolom-kolom numerik
numeric_df = df.select_dtypes(include='number')

# Menghitung korelasi antar kolom dalam dataframe numerik
correlation_matrix = numeric_df.corr()

# Membuat heatmap untuk memvisualisasikan korelasi antar kolom
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Antar Kolom')
plt.show()
```
